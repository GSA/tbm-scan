The code checks for the presence of the output file - `data.csv`. If that file is present, the scan has already been run at least once in the past, so we'll pick up where we left off with this scan by scanning from the latest date in that file up until the most recent date fbo.gov has in its FTP server. If `data.csv`doesn't exist, this must be the first time we've run the scan so we'll default to scanning the entire FY and then create the file. You can adjust how far back the scan goes in this case by adjusting date that's currently set to `'2018-10-01'` to whatever you like.

```python
from datetime import datetime, timedelta
import os

from tbm_scan import main

two_days_ago = (datetime.now() - timedelta(2)).strftime('%Y-%m-%d')

if not os.path.exists('data.csv'):
    print("Running this for the first time!")
    main(from_jupyter = True, start_date = '2018-10-01', end_date = two_days_ago, tbm_filter = True)
    print("All done!")
else:
    print("Picking up where we left off...")
    main(from_jupyter = True, tbm_filter = True)
    print("All done!")
```
